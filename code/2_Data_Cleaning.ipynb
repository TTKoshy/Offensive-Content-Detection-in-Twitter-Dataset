```python
from transformers import TFRobertaForSequenceClassification, RobertaTokenizer
import tensorflow as tf
import pandas as pd
import re
import emoji
from langdetect import detect, DetectorFactory
import os
```

```python
def is_just_emojis(text):
    return all(char in emoji.EMOJI_DATA for char in text)

def preprocess_text(text):
    text = re.sub(r"http\S+|www\S+|https\S+", "", text, flags=re.MULTILINE)
    text = re.sub(r"#\S+|@\S+", "", text)
    text = re.sub(r"[^A-Za-z\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return None if not text or is_just_emojis(text) else text

def is_english(text):
    try:
        return detect(text) == "en"
    except:
        return False
```

```python
base_path = "/content/drive/MyDrive/MRP_Offensive_Content_Detection/Twitter_Data/"
batch_size = 500
MAX_LENGTH = 512

for i in range(38, 39):
    dataset_name = f"Part_{i}_pre_processed.csv"
    DetectorFactory.seed = 0
    output_path = base_path + f"Part_{i}/" + f"Part_{i}_processed_new.csv"

    for chunk in pd.read_csv(f"{base_path}Part_{i}/{dataset_name}", chunksize=batch_size):
        chunk.dropna(subset=['text'], inplace=True)
        chunk['text'] = chunk['text'].apply(preprocess_text)
        chunk.dropna(subset=['text'], inplace=True)

        chunk['text'] = chunk['text'].apply(lambda x: ' '.join(x.split()[:MAX_LENGTH]))
        chunk = chunk[chunk['text'].apply(is_english)]

        # Append the processed chunk to the output file
        chunk.to_csv(output_path, mode='a', index=False, header=not pd.io.common.file_exists(output_path))
        print(f"Processed a batch of {len(chunk)} rows. Total saved to {output_path}.")

    print(f"Completed processing for dataset Part_{i}.")
```
